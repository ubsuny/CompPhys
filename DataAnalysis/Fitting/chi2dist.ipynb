```python
import scipy 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0., 100., 1000)
chi2 = {}
chi2ndf = {}
for dof in range(1, 10):
    chi2[dof] = scipy.stats.chi2.pdf(x, dof)
    chi2ndf[dof] = chi2[dof] / dof
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
for dof in range(1, 10):
    ax.plot(x/dof, chi2ndf[dof]*dof, label=f"NDOF={dof}")
ax.set_xscale("linear")
ax.set_xlabel(r"$\chi^2/$NDF")
ax.set_xlim([0, 10.])
ax.set_yscale("log")
ax.set_ylim([1.e-3, 10.])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
for dof in range(1, 10):
    ax.plot(x, chi2[dof], label=f"NDOF={dof}")
ax.set_xscale("linear")
ax.set_xlabel(r"$\chi^2$")
ax.set_xlim([0, 10.])
ax.set_yscale("log")
ax.set_ylim([1.e-3, 10.])
```
