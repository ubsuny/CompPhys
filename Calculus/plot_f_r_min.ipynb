```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.linspace(0.1,3,3000)
V = np.where(r < 1., 1., 0.)
```

```python
for b in np.linspace(0.5,1.5,5):
    plt.plot( r, 1 - np.power( b/r, 2) - V, label="%3.1f"%(b))
plt.ylim(-5,5)
plt.xlabel(r"$r$")
plt.ylabel(r"$f(r_{min})$")
plt.legend()
plt.axhline(color='k')
```
